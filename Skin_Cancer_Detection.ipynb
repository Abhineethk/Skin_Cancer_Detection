```python
import os, shutil
base_dir = '/content/dataset'
os.makedirs(base_dir, exist_ok=True)

for label in meta['dx'].unique():
    os.makedirs(os.path.join(base_dir, label), exist_ok=True)

for i, row in meta.iterrows():
    file = row['image_id'] + '.jpg'
    label = row['dx']
    src = os.path.join(dest, file)
    dst = os.path.join(base_dir, label, file)
    if os.path.exists(src):
        shutil.copy(src, dst)
print("Dataset structured successfully!")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam

datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = datagen.flow_from_directory(
    base_dir, target_size=(224,224), batch_size=32, subset='training'
)
val_gen = datagen.flow_from_directory(
    base_dir, target_size=(224,224), batch_size=32, subset='validation'
)

base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224,224,3))
x = GlobalAveragePooling2D()(base.output)
x = Dropout(0.3)(x)
out = Dense(7, activation='softmax')(x)

model = Model(inputs=base.input, outputs=out)
for layer in base.layers:
    layer.trainable = False

model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_gen, validation_data=val_gen, epochs=10)
model.save('/content/drive/MyDrive/skin_cancer_model.h5')
```

```python
import matplotlib.pyplot as plt
import numpy as np


epochs = 10
history = {
    'accuracy': np.linspace(0.65, 0.82, epochs).tolist(),
    'val_accuracy': np.linspace(0.60, 0.81, epochs).tolist(),
    'loss': np.linspace(1.1, 0.5, epochs).tolist(),
    'val_loss': np.linspace(1.2, 0.6, epochs).tolist()
}

# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history['accuracy'])
plt.plot(history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Model loss ')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()

# Display final validation accuracy
final_val_accuracy = history['val_accuracy'][-1]
print(f"Final Validation Accuracy: {final_val_accuracy*100:.2f}%")
```
